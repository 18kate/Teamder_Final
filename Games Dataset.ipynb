```python
def create_game(row):
    url = 'http://127.0.0.1:5050/api/games'
    body = {
        'name': row['name'],
        'has_rank': False
    }
    if str(row['desc_snippet']) == 'nan':
        pass
    else:
        body['description'] = row['desc_snippet']
    if str(row['image_url']) == 'nan':
        pass
    else:
        body['picture_url'] = row['image_url']
    
    print(body)
    
    responce = requests.post(url, json=body)
```
